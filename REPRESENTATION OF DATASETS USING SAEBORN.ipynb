```python
import seaborn as sns
import pandas as pd
pd.set_option("display.max_row", None)

dataset = sns.load_dataset("penguins")

#dataset=  pd.get_dummies(dataset)
dataset
```

```python
sns.lmplot(x = "bill_length_mm", y = "bill_depth_mm",data = dataset, hue = "species", markers = ["x", "o", "v"])
```

```python
sns.lmplot(x = "bill_length_mm", y = "bill_depth_mm",data = dataset, col = "sex", hue = "sex", markers = ["x", "o"])
```

```python
try: 
    dataset = dataset.drop(["species"], axis = 1)
    dataset = dataset.drop(["island"], axis = 1)
    dataset = dataset.drop(["sex"], axis = 1)
    crln = dataset.corr()
    sns.heatmap(crln)
    
except:
    print("sorry")
   

```

```python
sns.pairplot(dataset, hue = "sex")
```
